```python
from transformers import AutoTokenizer
```

```python
MODEL_NAME = "bert-base-uncased"
tokenizer =AutoTokenizer.from_pretrained(MODEL_NAME)
print(f" Using Tokenizer: (MODEL_NAME)")
# 4. Define Input Text Examples
# Example 1: Standard words
text_1= "Tokenization is fundamental for large language models."
#Example 2: Rare word, hyphenation, and specialized characters
#'Subword' is a common word, but 'unpredictable' often gets split.
text_2= "Unpredictable supercalifragilisticexpialidocious $1M"
```

```python
#5. Tokenization and Encoding
def tokenize_and_display(input_text):
  #1. Tokenize (convert text to the subword units)
  tokens_decoded= tokenizer.tokenize (input_text)
  print(f"\nInput Text: (input_text)'")
  print("-" * 50)
  print("A. Subword Tokens (Decoded):")
  print(tokens_decoded)
  print(f"Total Tokens: {len(tokens_decoded)}")

```

```python
  # Run the demo for the examples
  tokenize_and_display(text_1)
  tokenize_and_display(text_2)
```
